## Task 1: Conceptual Understanding

### 1. Difference between "Love" and "love" in NLP
In NLP, "Love" and "love" are treated as different words because of case sensitivity. 
This increases vocabulary size unnecessarily. To avoid this, we convert all text to lowercase during preprocessing.

---

### 2. What happens if stopwords are not removed?
If stopwords are not removed:
- The dataset will contain unnecessary common words like "is", "the", "and"
- It increases processing time
- It may reduce model performance due to noise

---

### 3. When removing stopwords can be harmful
1. Sentiment Analysis:
   Example: "I am not happy"
   Removing "not" changes meaning completely.

2. Question Answering:
   Example: "What is your name?"
   Removing "what" removes the intent of the question.

---

### 4. Stemming vs Lemmatization

Stemming:
- Cuts words to root form
- Example: "running" → "run"
- Faster but less accurate

Lemmatization:
- Uses dictionary-based approach
- Returns meaningful root word
- More accurate but slower



## Task 2: Preprocessing Function

We create a function to clean and normalize text.

Steps performed:
- Convert to lowercase
- Remove URLs and emails
- Remove numbers
- Remove special characters and emojis
- Handle repeated characters
- Remove extra spaces
- Remove short words (<=2 except "no", "not")

```python
import re

def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""
    
    text = text.lower()

    # Remove URLs and emails
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove special characters & emojis
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Handle repeated characters
    text = re.sub(r'(.)\1+', r'\1', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short words
    tokens = [w for w in tokens if len(w) > 2 or w in ['no', 'not']]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence


print(preprocess_text(input()))
```

## Task 3: Stress Testing

In this step, we test our preprocessing function using diverse and noisy text inputs.

The goal is to ensure that the function works correctly on real-world data such as:
- Sentences with emojis
- Numbers and percentages
- URLs and links
- Repeated characters
- Mixed uppercase and lowercase text

This helps validate the robustness of our preprocessing pipeline.

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for i in sentences:
    tokens, clean = preprocess_text(i)
    
    print("Original Text:", i)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    print("-" * 50)
```

## Task 4: Token Analytics

In this step, we analyze the cleaned tokens generated from our preprocessing function.

We compute:
- Total number of tokens in each sentence
- Number of unique tokens
- Average length of tokens

This helps us understand how much useful information is retained after cleaning.

```python
import numpy as np

for s in sentences:
    tokens, _ = preprocess_text(s)

    # Total tokens
    total_tokens = len(tokens)

    # Unique tokens
    unique_tokens = len(set(tokens))

    # Average token length
    if len(tokens) > 0:
        avg_length = sum(len(word) for word in tokens) / len(tokens)
    else:
        avg_length = 0

    print("Sentence:", s)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
    print("-" * 50)
```

## Task 5: Frequency Analysis

In this step, we analyze the frequency of words across all processed sentences.

We combine all tokens into a single list and use the Counter class to:
- Identify the most frequent words
- Identify the least frequent words

This helps in understanding common patterns and rare words in the dataset.

```python
from collections import Counter

all_tokens = []

# Collect all tokens
for s in sentences:
    tokens, _ = preprocess_text(s)
    all_tokens.extend(tokens)

# Count frequency
counter = Counter(all_tokens)

# Top 10 most frequent words
print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nLeast 5 Frequent Words:")
print(counter.most_common()[-5:])

print("\nTop 10 Words:")
for word, freq in counter.most_common(10):
    print(f"{word} → {freq}")

print("\nLeast 5 Words:")
for word, freq in counter.most_common()[-5:]:
    print(f"{word} → {freq}")
```

## Task 6: Full Pipeline

In this step, we build a complete NLP preprocessing pipeline.

The pipeline:
- Takes a list of raw text inputs
- Applies preprocessing to each sentence
- Collects all tokens
- Stores cleaned sentences

This function makes our preprocessing reusable and scalable for real-world applications.

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

result = full_pipeline(sentences)

print("All Tokens:")
print(result["tokens"])

print("\nCleaned Sentences:")
for s in result["clean_sentences"]:
    print(s)
```

To make our preprocessing function robust, we handle edge cases such as:
